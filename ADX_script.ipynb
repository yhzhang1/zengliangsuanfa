```python
def TrueRange(high, low, preClose):
    return max([abs(high-low), abs(high-preClose), abs(preClose-low)])

def DX_inc(plusDI, minusDI):
    return 100 * abs(plusDI - minusDI) / (plusDI + minusDI)

def ADX(high, low, close, n = 14):
    deltahigh = np.full_like(high, fill_value=np.nan)
    deltalow = np.full_like(high, fill_value=np.nan)
    plusdm = np.full_like(high, fill_value=np.nan)
    minusdm = np.full_like(high, fill_value=np.nan)
    admplus = np.full_like(high, fill_value=np.nan)
    admminus = np.full_like(high, fill_value=np.nan)
    tr = np.full_like(high, fill_value=np.nan)
    atr = np.full_like(high, fill_value=np.nan)
    plusdi = np.full_like(high, fill_value=np.nan)
    minusdi = np.full_like(high, fill_value=np.nan)
    dx = np.full_like(high, fill_value=np.nan)
    adx = np.full_like(high, fill_value=np.nan)
    
    
    for i in range(len(high)):
        if i == 0:
            pass
        else:
            deltahigh[i] = high[i] - high[i-1] # 更新deltahigh
            deltalow[i] = low[i-1] - low[i] # 更新deltalow
            tr[i] = TrueRange(high[i], low[i], close[i-1]) # 更新truerange
            # 更新 plusdm minusdm
            if (deltahigh[i] < 0 and deltalow[i] < 0) or deltahigh[i] == deltalow[i]:
                plusdm[i] = 0
                minusdm[i] = 0
            elif deltahigh[i] > deltalow[i]:
                plusdm[i] = deltahigh[i]
                minusdm[i] = 0
            elif deltahigh[i] < deltalow[i]:
                plusdm[i] = 0
                minusdm[i] = deltalow[i]
            if i == 1:
                # 更新第一个 adm
                
                admplus[i] = plusdm[i]
                admminus[i] = minusdm[i]
                
                # 更新第一个 atr
                atr[i] = tr[i]
            else:
                # 更新第二个及以后的 adm
                
                admplus[i] = (plusdm[i] + (n-1)*admplus[i-1]) / n
                admminus[i] = (minusdm[i] + (n-1)*admminus[i-1]) / n

                # 更新第二个及以后的 atr
                atr[i] = (tr[i] + (n-1)*atr[i-1]) / n
            # 更新 di
            plusdi[i] = 100 * admplus[i] / atr[i]
            minusdi[i] = 100 * admminus[i] / atr[i]
            # 更新 dx
            dx[i] = DX_inc(plusdi[i], minusdi[i])
            
            if i == 1:
                adx[i] = dx[i]
                print(dx[i])
            else:
                adx[i] = ((n-1) * adx[i-1] + dx[i]) / n
            
    return adx
```
