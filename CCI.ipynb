```python
def CCI_inc(typical_price, avg_typical_price, avg_deviation):
    return (typical_price - avg_typical_price)/(0.015*avg_deviation)

def CCI(high, low, close, timeperiod=14):
    typical_price = np.full_like(high, fill_value=np.nan)
    avg_typical_price = np.full_like(high, fill_value=np.nan)
    mean_deviation = np.full_like(high, fill_value=np.nan)
    cci = np.full_like(high, fill_value=np.nan)
    
    for i in range(len(high)):
        typical_price[i] = (high[i] + low[i] + close[i]) / 3
        if i >= timeperiod-1:
            if i == timeperiod-1:
                avg_typical_price[i] = typical_price[i-timeperiod+1:i+1].mean()
            elif i > timeperiod-1:
                avg_typical_price[i] = avg_typical_price[i-1] - typical_price[i-timeperiod]/timeperiod + \
                                        typical_price[i]/timeperiod
            mean_deviation[i] = abs(typical_price[i-timeperiod+1:i+1] - avg_typical_price[i]).mean()
            cci[i] = CCI_inc(typical_price[i], avg_typical_price[i], mean_deviation[i])
            
    return cci
    
```

```python
import numpy as np
size = 1000
high = np.random.random(size)
low = np.random.random(size)
close = np.random.random(size)
```
