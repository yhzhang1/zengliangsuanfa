```python


def KDJ(high, low, close, timeperiod=14):
    highest_high = np.full_like(high, fill_value=np.nan)
    lowest_low = np.full_like(high, fill_value=np.nan)
    rsv = np.full_like(high, fill_value=np.nan)
    k = np.full_like(high, fill_value=np.nan)
    d = np.full_like(high, fill_value=np.nan)
    j = np.full_like(high, fill_value=np.nan)
    k_n = 9
    d_n = 9
    
    for i in range(len(high)):
        if i >= timeperiod - 1:
            lowest_low[i] = low[i+1-timeperiod:i+1].min()
            highest_high[i] = high[i+1-timeperiod:i+1].max()
            rsv[i] = (close[i] - lowest_low[i])/(highest_high[i] - lowest_low[i]) * 100
            if i == timeperiod - 1 + k_n - 1:
                k[i] = rsv[i+1-k_n:i+1].mean()
            elif i > timeperiod - 1 + k_n - 1:
                k[i] = k[i-1] - rsv[i-k_n]/k_n + rsv[i]/k_n
                if i >= timeperiod - 1 + k_n - 1 + d_n - 1:
                    if i == timeperiod - 1 + k_n - 1 + d_n - 1:
                        d[i] = k[i+1-d_n:i+1].mean()
                    elif i > timeperiod - 1 + k_n - 1 + d_n - 1:
                        d[i] = d[i-1] - k[i-d_n]/d_n + k[i]/d_n
                    j[i] = 3*k[i] - 2*d[i]
                    
    return k, d, j
```

```python
import numpy as np

size = 100
high = np.random.random(size)
low = np.random.random(size)
close = np.random.random(size)
```

```python
KDJ(high, low, close)
```
