```python


def TEMA(close, n=30):
    ema1 = np.full_like(close, fill_value=np.nan)
    ema2 = np.full_like(close, fill_value=np.nan)
    ema3 = np.full_like(close, fill_value=np.nan)
    tema = np.full_like(close, fill_value=np.nan)

    
    k = 2/(n+1)
    for i in range(len(close)):
        if i == 0:
            # 更新第一个ema1 ema2 ema3
            ema1[i] = close[i]
            ema2[i] = ema1[i]
            ema3[i] = ema2[i]
        else:
            ema1[i] = ((n-1)/(n+1))*ema1[i-1] + (2/(n+1))*close[i]
            ema2[i] = ((n-1)/(n+1))*ema2[i-1] + (2/(n+1))*ema1[i]
            ema3[i] = ((n-1)/(n+1))*ema3[i-1] + (2/(n+1))*ema2[i]
            
        tema[i] = TEMA_inc(ema1[i], ema2[i], ema3[i])    
        
    return tema
    

    

def TEMA_inc(ema1, ema2, ema3):
    return 3*ema1 - 3*ema2 + ema3
```
